```python
df_new = df.reindex(columns=['日付','年', '月', '日' ,'平均気温(℃)' ])
df_new
```

```python
str(df.index[0])
```

```python
# dfを10個ずつに分割

# N = 10 
# splited_df = [df[i:i+N] for i in range(0, len(df), N)]

# 年毎に分割

split_day = [0]

for i in range(len(df)):
    if df.iat[i,1] == 12 and df.iat[i,2] == 31:
        split_day.append(i+1)

splited_df = []
splited_df = [df[split_day[i]:split_day[i+1]] for i in range(len(split_day)-1)]
              
splited_df[0]
```

```python
splited_df[0].iat[0,0]
```

```python
s_df = []

for i in range(len(splited_df)):
    s = splited_df[i]
    year = s.iat[0,0]
    splited_df[i] = s.rename(columns={'平均気温(℃)': year})
    
    s_df.append(splited_df[i].drop(splited_df[i].columns[[0]], axis=1))
    
s_df[1][55:65]
```

```python
s_df[1].columns.values[2]
```

```python
from datetime import datetime as dt
```

```python
tmp_date = str(2020) + '-02-01 00:00:00'
tmp_date = dt.strptime(tmp_date, '%Y-%m-%d %H:%M:%S')
print(tmp_date)
```

```python
# うるう年を追加したい
# うるう年ではない年でも2/29を追加し、前日と翌日の平均としておく

for k in range(len(s_df)):
    s = s_df[k]
    for i in range(len(s)):
        if s.iat[i,0] == 2 and s.iat[i,1] == 28 and s.iat[i+1,0] == 3:
            year = str(s.columns.values[2])+'-02-29 00:00:00'
            tmp_date = dt.strptime(year, '%Y-%m-%d %H:%M:%S')
            print(tmp_date)
            df_front = s[0:i+1]
            df_mid = pd.DataFrame({'月': [2],
                                   '日': [29]},
                                  index = [tmp_date])
            df_back = s[i+1:]
            df_tmp = pd.concat([df_front,df_mid])
            df_tmp = pd.concat([df_tmp,df_back])
            s_df[k] = df_tmp
            
s_df[1][55:65]
```
